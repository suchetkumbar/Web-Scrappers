# Finding Elements in a Web Page

Finding elements by class and by ID

```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.tutorialsfreak.com/")
web
```

```python
web.content
```

```python
soup = BeautifulSoup(web.content, "html.parser")
print(soup.prettify)
```

```python
soup.find_all("p")
```

```python
soup.find_all("a")
```

```python
class_data = soup.find("div",class_="app-container") # find by class name
print(class_data)

#class_data.find_all("p") # nesting ops
```

```python
id_data = soup.find("div",id="app")
print(id_data)
```

# Extracting Texts from Tags

```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.tutorialsfreak.com/")
web
```

```python
soup = BeautifulSoup(web.content, "html.parser")
```

```python
lines = soup.find_all("p")
lines
```

```python
# extracting content from the list generated above

for l in lines:
  print(l.text)
```

```python
s = soup.find("div",class_="bg-white border-0 p-3 card-footer")
print(s)
```

```python
l1 = s.find_all("a")
l1
```

```python
for l in l1:
  print(l.text)
```

```python
s1 = soup.find("a", class_="tf-button setup-btn me-sm-3 me-2 exp-all-btn") # specify the detailed class name
s1
#s1.text
```

# Extracting Links in a Web Page using Python

```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.tutorialsfreak.com/")
web
```

```python
soup = BeautifulSoup(web.content, "html.parser")
soup
```

```python
for i in soup.find_all("a"):
  print(i.get("href"))
```

# Extracting Image Information in Web Page using Python

```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.tutorialsfreak.com/")
web
```

```python
soup = BeautifulSoup(web.content, "html.parser")
print(soup.prettify())
```

```python
img = soup.find_all("img")
img
```

```python
for i in img:
  print(i.get("src"))
  print(type(i.get("src")))
  #print(i.get('alt'))
```
