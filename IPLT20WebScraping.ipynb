```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.iplt20.com/auction/2022"
r = requests.get(url)
r
```

```python
soup = BeautifulSoup(r.text,"lxml")
print(soup.prettify())
```

```python
table = soup.find("table",class_="ih-td-tab w-100 auction-tbl")
title = table.find_all("th")
```

```python
print(table)
```

```python
print(title)
```

```python
header = []
for i in title:
  name = i.text
  header.append(name)
print(header)
```

```python
df = pd.DataFrame(columns = header)
```

```python
print(df)
```

```python
rows = table.find_all("tr")
print(rows) # need to skip first row since it consists of the table headers
```

```python
for i in rows[1:]:
  data = i.find_all("td")
  #print(data)
  row = [tr.text for tr in data]
  l = len(df)
  df.loc[l] = row
print(df)
```

```python
df.to_csv("IPL Auction stats.csv") # saves to an excel file
```
